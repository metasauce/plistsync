# Plex

The following sections describe how to configure and use the *Plex* integration with `plistsync`.

## Getting started

If you have not done so already make sure to enable and configure the Plex integration in your configuration file. If you are running your own server, make sure to change the server
url.

```yaml
# ./config/config.yaml
# Add the following to your config
plex:
    enabled: true
    # Usually you want to use one of, if both are defined url takes precedence
    server_url: "http[s]://[url]:[port]" 
    server_name: "my_server"
```

To use the Plex service within plistsync, you need to authorize `plistsync` to access your Plex account. Run `plistsync plex auth` and follow the prompts to complete the authorization process. This will save the token used for authentication next to the configuration file. 

```{typer} plistsync.services.plex.authenticate:plex_cli
---
prog: plistsync plex auth
---
```


## Library

To access your Plex server, you must select a *section*. In the Plex frontend, this is referred to as a *library*. In this documentation, however, we use the term *section* to align with the Plex backend specification.

A section represents a library section within your Plex media library. With the default Plex configuration, this section is typically named `Music`.


````python
from plistsync.services.plex.collection import PlexLibrarySectionCollection

collection = PlexLibrarySectionCollection(
    "Music",
    # server_name = "my_plex_server",
    # server_url = "http://localhost:3200"
    # if you run multiple server you can select a specific one using server_url
    # or server_name. If none is given we use the default one from the config
    # or will try to resolve the server url via the name
)
````

### Loading all playlists in your section

Using your `PlexLibrarySectionCollection` you can load all available playlists:

````python
for pl in collection.playlists:
    print(pl)
````

This returns a list of `PlexPlaylistCollection`.
