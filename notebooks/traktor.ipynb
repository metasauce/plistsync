# Traktor read playlist

```python
from pathlib import Path

nml_collection = Path("/mnt/media/music/collection.nml")
```

```python
from plistsync.services.traktor import NMLCollection

nml = NMLCollection(nml_collection)
```

```python
for pl in nml.playlists():
    print(pl.name)
```

```python
pl = nml.playlist("rave waiting room")
```

```python
path_rewrite = [("//SYNC/", "/mnt/media/music/")]

paths = []
tracks = [track for track in pl]
for track in tracks:
    path = track.path
    for old, new in path_rewrite:
        path = str(path).replace(old, new)
    paths.append(path)
```

```python
paths
```

```python
# Create zip from paths
import os
import zipfile
from io import BytesIO

zip_buffer = BytesIO()
with zipfile.ZipFile(zip_buffer, "w") as zip_file:
    for path in set(paths):
        zip_file.write(path, os.path.basename(path))

zip_buffer.seek(0)
# Save zip to file
with open("/mnt/media/music/rave_waiting_room.zip", "wb") as f:
    f.write(zip_buffer.getvalue())
```
