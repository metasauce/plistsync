# Spotify usage guide

This notebook demonstrates how to use the Spotify service in the `plistsync` library to interact with Spotify playlists and libraries.


## Getting started

If you haven't already, make sure to set the `spotify` configuration in your `config/config.yaml` file.

```yaml
# ./config/config.yaml
...
spotify:
  enabled: true 
  client_id: your_spotify_client_id
  client_secret: your_spotify_client_secret
```

You can check if this is set up correctly by running:

```python
from plistsync.config import Config
Config().spotify.__dict__
```

Next you need to authenticate with Spotify. You can do this by running 

```cmd
plistsync auth spotify
```

## Playlists

As the package name suggests, `plistsync` is primarily focused on syncing playlists. 


### Loading a specific playlist by ID

You may load a specific playlist by its Spotify ID using the `SpotifyPlaylistCollection.from_id` method. You can find the
id of the playlist in the share link of the playlist, e.g. in `https://open.spotify.com/playlist/0psGNXS20icKoZPeAJJ5sg`, the id is `0psGNXS20icKoZPeAJJ5sg`.

````python
from plistsync.services.spotify import SpotifyPlaylistCollection

litt_dnb = await SpotifyPlaylistCollection.from_id("0psGNXS20icKoZPeAJJ5sg")

# Alternatively, you may use asyncio.run to run the method in a synchronous context:
import asyncio
litt_dnb = asyncio.run(SpotifyPlaylistCollection.from_id("0psGNXS20icKoZPeAJJ5sg"))

# This also works by name if you have saved the playlist to your library:
data = SpotifyPlaylistCollection.from_id("<playlist id or name>")
````

### Loading all playlists in your library

When authenticating with Spotify, you were asked to give permission to read your library. This allows you to load all playlists in your library using the `SpotifyLibraryCollection.playlists` method. Depending on the number of playlists in your library, this may take a while.

````python
from plistsync.services.spotify import SpotifyLibraryCollection

user_playlists = SpotifyLibraryCollection().playlists
````

### Accessing playlist properties

When we fetch the playlist data using either of the methods above, we get a `SpotifyPlaylistCollection` object. This object has several properties that we can access:

- `id`: The unique identifier for the playlist.
- `name`: The name of the playlist.
- `__iter__`: Returns an iterator over the tracks in the playlist.


If you want to use the returned response data directly, you can access it via the `data` property.

````python
litt_dnb.name
````

````python
litt_dnb.id
````

````python
# See https://developer.spotify.com/documentation/web-api/reference/get-playlist
litt_dnb.data["owner"]["display_name"]
````
